```python
import pandas as pd
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data = pd.read_csv(url)
print(data)
month = data["month_number"]
profits = data["total_profit"]
plt.plot(month, profits)
plt.xlabel("month")
plt.ylabel("profits")
plt.title("Total Profit by Month")
plt.grid(True)
plt.marker = 'o'
plt.show()
```
